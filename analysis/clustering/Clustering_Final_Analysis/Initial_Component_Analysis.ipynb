# I.5- Survival Plots

```r
surv_object <- Surv(time = df_w_correlates$os, event = df_w_correlates$os_status)
```

```r
set_notebook_plot_size(16,12)
fit1 <- survfit(surv_object ~ initial_predicted_component, data = df_w_correlates)

ggsurvplot(fit1, data = df_w_correlates, pval = TRUE,risk.table = TRUE,ggtheme = theme_minimal(),
 risk.table.y.text.col = T,
  risk.table.y.text = T,break.time.by = 1,title= "Survival by Initial component",palette=val)
```

```r
set_notebook_plot_size(8,3.6)
fit_all <- survfit(surv_object ~1, data = df_w_correlates)
ggsurvplot(fit_all, data = df_w_correlates, risk.table = TRUE,ggtheme = theme_minimal(),
 risk.table.y.text.col = T,
  risk.table.y.text = T,break.time.by = 1,title="Overall Survival",palette="grey45")
```

# I.6- Clinical categorical correlates for each component VS others

<a id='catcorr'></a>

```r
set_notebook_plot_size(50,3)
tmp <-df_w_correlates
list_component <- levels(factor(df_w_correlates$initial_predicted_component))
categlist <- list()
for (col in binary_correlates){
    for (i in list_component){
    tmp$comparison <- ifelse((tmp$initial_predicted_component==i),i,"Others") 
    categlist[[i]]<- ggplot(tmp, aes(x = factor(comparison), fill = factor(tmp[,col]))) + 
                geom_bar(position = "fill") +scale_y_continuous(labels = scales::percent)+xlab("")+
                scale_fill_manual(values = val)+labs(fill=col) +theme(axis.text.x = element_text(angle = 35, hjust = 1))

    }
    grid.arrange(grobs=categlist,ncol=length(list_component))
}
```

# I.8- Survival Plot for each component VS Others

# A) Let's see if there is a difference in the two t(8,21) subdivisions

```r
tmp1 <- tmp[is.element(tmp$initial_predicted_component,6:7),]
tmp1$comparison <- ifelse((tmp1$initial_predicted_component==6),"First_subdivision_t_8_21","Second_subdivision_t_8_21") 
surv_object <- Surv(time = tmp1$os, event = tmp1$os_status)
fit <- survfit(surv_object ~ comparison, data = tmp1)
set_notebook_plot_size(10,6)
ggsurvplot(fit, data = tmp1, pval = TRUE,risk.table = TRUE,ggtheme = theme_minimal(),
 risk.table.y.text.col = T,
  risk.table.y.text = T,break.time.by = 1,title= "Survival by Initial component",palette=val)
options(warn = -1)
ggplot(tmp1,aes(x=factor(comparison),y=tmp1[,"wbc"],fill=comparison)) +
                geom_boxplot(aes(fill = factor(comparison)),outlier.shape = NA) +  ylab("wbc")+scale_fill_manual(values=c("grey45","#e79f00"))+scale_y_continuous(limits = quantile(tmp[,"wbc"], c(0.1, 0.9),na.rm = T))+
        theme(axis.text.x = element_text(angle = 35, hjust = 1))+xlab("")+ggtitle("Comparison of white blood cells for those 2 initial Components") +stat_compare_means()
```

```r
tmp1<- merge(df_initial[correlates],df_all_components[c('initial_predicted_component','t_8_21')],by=0)
tmp1 <- tmp1[is.element(tmp1$initial_predicted_component,6:7) & tmp1$t_8_21==1,]
tmp1$comparison <- ifelse((tmp1$initial_predicted_component==6),"First_subdivision_t_8_21","Second_subdivision_t_8_21") 
surv_object <- Surv(time = tmp1$os, event = tmp1$os_status)
fit <- survfit(surv_object ~ comparison, data = tmp1)
set_notebook_plot_size(10,6)
ggsurvplot(fit, data = tmp1, pval = TRUE,risk.table = TRUE,ggtheme = theme_minimal(),
 risk.table.y.text.col = T,
  risk.table.y.text = T,break.time.by = 1,title= "Survival by Initial component with translocation",palette=val)
options(warn = -1)
ggplot(tmp1,aes(x=factor(comparison),y=tmp1[,"wbc"],fill=comparison)) +
                geom_boxplot(aes(fill = factor(comparison)),outlier.shape = NA) +  ylab("wbc")+scale_fill_manual(values=c("grey45","#e79f00"))+scale_y_continuous(limits = quantile(tmp[,"wbc"], c(0.1, 0.9),na.rm = T))+
        theme(axis.text.x = element_text(angle = 35, hjust = 1))+xlab("")+ggtitle("Comparison of white blood cells for those 2 initial Components with translocation")+stat_compare_means()

```
